## 실제 코드
- `get_page_data(url)`
  - 개별 페이지의 데이터를 추출하는 함수
  - 주어진 URL에서
    - 메타데이터(잡지명, 발행일, 기사제목, 기사형태, 필자) 추출
    - 본문 내용(content) 추출
    - URL 정보 저장
  - 추출한 데이터를 정리해서 dictionary 형태로 반환
  - 페이지가 없거나 오류가 있으면 None 반환
- `scrape_articles()`
  - 전체 스크래핑 과정을 관리하는 메인 함수
  - 다음과 같은 작업을 수행:
    - 호수별로 순차적으로 접근 (0010 → 0020 → 0030...)
    - 각 호수 내에서 기사번호 순서대로 접근 (0001 → 0010 → 0020...)
    - `get_page_data` 함수를 호출해서 각 페이지의 데이터 수집
    - 수집된 모든 데이터를 하나의 리스트로 모음
    - 최종적으로 DataFrame으로 변환하여 TSV 파일로 저장

```python
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup


def get_page_data(url):
    try:
        req = requests.get(url)

        # 404 에러 확인
        if req.status_code == 404:
            return None

        # Requests로 가져온 HTML 문서를 파싱하여 BeautifulSoup 객체 생성
        soup = BeautifulSoup(req.text, 'html.parser')

        # 메타데이터 추출
        meta_section = soup.find('section', class_='section-meta')
        if not meta_section:
            return None

        meta_data = {}
        for item in meta_section.find_all('div', class_='item'):
            title = item.find('div', class_='tit').text
            content = item.find('div', class_='cont').text.strip()
            # 필자 정보에서 공백 제거
            if title == '필자':
                content = content.replace(' ', '').replace('\n', '').replace('&nbsp;', '')  # 공백, \n, &nbsp; 제거
            meta_data[title] = content

        # 콘텐츠 추출
        content_div = soup.find('div', id='cont_view')
        if content_div:
            content_texts = content_div.find_all('div', style="text-align:justify;word-break:break-all;")
            content = '\\n'.join([text.text.strip().replace('\t', ' ').replace('\n', ' ') for text in content_texts])
        else:
            content = ''

        # 칼럼 순서에 맞게 데이터 정리
        ordered_data = {
            '잡지명': meta_data.get('잡지명', ''),
            '발행일': meta_data.get('발행일', ''),
            '기사제목': meta_data.get('기사제목', ''),
            '기사형태': meta_data.get('기사형태', ''),
            '필자': meta_data.get('필자', ''),  # 필자 정보 추가
            'content': content,
            'url': url
        }

        return ordered_data

    except Exception as e:
        print(f"오류 발생: {str(e)}")
        return None

def scrape_articles():
    base_url = 'https://db.history.go.kr/modern/level.do?levelId=ma_013_{}_{}'
    all_data = []
    issue_num = 10  # 시작 호수

    while True:  # 호수 반복
        issue_str = str(issue_num).zfill(4)  # 0010, 0020 등의 형태로 변환
        article_id = 1  # 각 호수별 첫 기사 ID를 1로 시작
        first_article = True
        continuous_fails = 0  # 연속 실패 횟수 추가

        print(f"\n=== {issue_str}호 처리 시작 ===")

        while article_id <= 500:  # article_id가 500까지
            article_str = str(article_id).zfill(4)
            url = base_url.format(issue_str, article_str)

            print(f"URL 처리 중: {url}")

            # 각 article_id마다 3번까지 시도
            success = False
            retry_count = 0
            while retry_count < 3 and not success:
                data = get_page_data(url)
                if data:
                    all_data.append(data)
                    success = True
                    continuous_fails = 0  # 성공하면 연속 실패 횟수 초기화
                else:
                    retry_count += 1
                    print(f"실패 횟수: {retry_count}")
                time.sleep(1)

            if not success:
                continuous_fails += 1
                if continuous_fails >= 3:  # 3번 연속으로 실패하면
                    print(f"{issue_str}호에 더 이상 기사가 없습니다.")
                    break  # 현재 호수 종료

            # 다음 article ID 설정
            if first_article:  # 첫 번째 기사(1)였다면 다음은 10
                article_id = 10
                first_article = False
            else:  # 그 이후로는 10씩 증가
                article_id += 10

        # 다음 호수로 이동
        issue_num += 10
        print(f"{issue_num}호로 이동합니다.")

        # 0750호 이후로 종료
        if issue_num > 30:  # 테스트만 할 경우 숫자를 낮게 설정 (가령 3호까지만 해 보려면 30으로 설정)
            print("마지막 호수까지 처리 완료")
            break

    # 결과를 DataFrame으로 변환
    df = pd.DataFrame(all_data)
    print(f"\n총 수집된 기사 수: {len(all_data)}")

    # TSV 파일로 저장
    output_filename = '개벽데이터_스크래핑.tsv'
    df.to_csv(output_filename, index=False, encoding='utf-8-sig', sep='\t', quoting=1, quotechar='"')
    print(f"데이터가 {output_filename}에 저장되었습니다.")

    return df

# 실행
print("스크래핑을 시작합니다...")
scraped_data = scrape_articles()
```
